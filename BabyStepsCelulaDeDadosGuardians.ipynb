```python
import pandas as pd
import matplotlib.pyplot as plt

urlDados = 'https://raw.githubusercontent.com/elasComputacao/raio-x-dados/main/data/dados-processados/notas.csv'

dados = pd.read_csv(urlDados)

periodos_de_ingresso = dados.groupby('periodo_ingresso')

media_periodos_de_ingresso = periodos_de_ingresso.mean().round(2)
media_periodos_de_ingresso = media_periodos_de_ingresso.drop('Unnamed: 0', axis=1)

plt.figure(figsize=(10, 6))
media_periodos_de_ingresso.plot(kind='line', color='skyblue')
plt.title('Mdia por Perodo de Ingresso')
plt.xlabel('Perodo de Ingresso')
plt.ylabel('Mdia')
plt.xticks(rotation=45)
plt.ylim(500, max(media_periodos_de_ingresso['media'])+100)

plt.show()
```
